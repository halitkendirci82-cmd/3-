1. 서론 (Introduction)
분석 목적과 데이터셋 설명을 정리한다. monks-problems-2 데이터셋은 범주형 변수와 타겟(class)으로 구성된 인공 데이터셋이다.

2. 데이터 로드
데이터를 불러와서 기본 구조를 확인한다.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

data = fetch_openml(name='monks-problems-2', as_frame=True)
df = data.frame
df.head()
```

2.1 데이터 명세 확인
데이터 크기와 변수 정보를 확인한다.

```python
df.shape
df.info()
```

2.2 결측치 분석
결측치 여부를 확인한다.

```python
df.isnull().sum()
```

2.3 기초 통계량
변수들의 기본 통계치를 확인한다.

```python
df.describe()
```

2.4 데이터 정제
중복 데이터를 제거한다.

```python
df = df.drop_duplicates()
df.shape
```

3. 변수별 분석
데이터는 대부분 범주형 변수로 구성되어 있다. 변수별 분포를 확인한다.

```python
for col in df.columns:
    print(df[col].value_counts())
sns.countplot(x='class', data=df)
plt.show()
```

3.1 파생 변수 생성
새로운 변수를 만들어 추가 분석을 시도한다.

```python
df['new_feature'] = df.iloc[:,0].astype(str) + "_" + df.iloc[:,1].astype(str)
df['new_feature'].value_counts()
```

4. 상관관계 및 관계 분석
변수 간 관계를 시각화한다.

```python
df_encoded = df.copy()
for col in df_encoded.columns:
    df_encoded[col] = df_encoded[col].astype('category').cat.codes

plt.figure(figsize=(8,6))
sns.heatmap(df_encoded.corr(), cmap='coolwarm')
plt.show()
```

4.2 Boxplot
타겟(class)에 따른 변수 분포를 확인한다.

```python
sns.boxplot(x='class', y=df_encoded.columns[0], data=df_encoded)
plt.show()
```

4.3 Scatter plot
변수 간 관계를 시각화한다.

```python
sns.scatterplot(
    x=df_encoded.columns[0],
    y=df_encoded.columns[1],
    hue="class",
    data=df_encoded
)
plt.show()

```
